```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
from sklearn.utils import shuffle
```

```python
df_full = pd.read_csv("DG_out.csv")
```

```python
df_less = pd.read_csv("BG.csv")
```

```python
df_full = df_full.drop(["tx_hash","in_malicious","out_malicious","is_malicious","all_malicious"],axis = 1)
df_full.head()
```

```python
df_less = df_less.drop(["tx_hash","in_malicious","out_malicious","is_malicious","all_malicious"],axis = 1)
df_less.head()
```

```python
smote = SMOTE(sampling_strategy=1)  # You can adjust the sampling_strategy as needed
df_less_balanced, _ = smote.fit_resample(df_less, df_less['out_and_tx_malicious'])
```

```python
df_less_balanced
```

```python
new_1_labels = df_less_balanced[df_less_balanced['out_and_tx_malicious'] == 1]
```

```python
new_1_labels
```

```python
df_full = pd.concat([df_full, new_1_labels], axis=0)
```

```python
df_full
```

```python
df_full = shuffle(df_full, random_state=42)
df_full
```

```python
count = 0
for i in df_full["out_and_tx_malicious"]:
    if i == 1:
        count+=1
print(count)
```
